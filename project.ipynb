```python
import pandas as pd

df["datetime"] = pd.to_datetime(df["datetime"],errors = "coerce")
df["date posted"] = pd.to_datetime(df["date posted"], errors = "coerce")

df["latitude"] = pd.to_numeric(df["latitude"] , errors = 'coerce')

df["longitude"] = pd.to_numeric(df["longitude"], errors = 'coerce')

df["duration (seconds)"] = pd.to_numeric(df["duration (seconds)"], errors = 'coerce')

# Convert the 'duration (hours/min)' column to string before using .str accessor
df['duration (hours/min)'] = df['duration (hours/min)'].astype(str)

# Now extract the number from the 'duration (hours/min)' column
df['duration (hours/min)'] = df['duration (hours/min)'].str.extract('(\\d+)', expand=False)

# Convert the extracted values to numeric
df["duration (hours/min)"] = pd.to_numeric(df["duration (hours/min)"], errors='coerce')


df["city"] = df["city"].astype("category")
df["state"] = df["state"].astype("category")
df["shape"] = df["shape"].astype("category")
df["country"] = df["country"].astype("category")
df["comments"] = df["comments"].astype("category")
```

```python
# we see that datetime has high number of NaN values so we need to drop it
# state, country,shape comments are categorical so we will use fillna
# numerical datas with NaN will get replaced with the mean or median values
```

```python
# This is Data Cleaning

df.dropna(subset=["datetime"], inplace=True)  

for col in ["state", "country", "shape", "comments"]:
    if "Unknown" not in df[col].cat.categories:  # Only add if not already present
        df[col] = df[col].cat.add_categories("Unknown")
    df[col] = df[col].fillna("Unknown")  # Assign back instead of inplace=True

df["duration (seconds)"] = df["duration (seconds)"].fillna(df["duration (seconds)"].median())
df["duration (hours/min)"] = df["duration (hours/min)"].fillna(df["duration (hours/min)"].median())
df["latitude"] = df["latitude"].fillna(df["latitude"].median())
```

```python
import pandas as pd
import numpy as np
from geopy.distance import geodesic
from sklearn.preprocessing import StandardScaler

# Ensure datetime column exists and is converted to datetime if necessary
if 'datetime' in df.columns:
    df['datetime'] = pd.to_datetime(df['datetime'], errors='coerce')
else:
    raise ValueError("The 'datetime' column is missing.")

# 1. Extracting date-time components (if not already done)
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day'] = df['datetime'].dt.day
df['hour'] = df['datetime'].dt.hour
df['day_of_week'] = df['datetime'].dt.dayofweek  # 0=Monday, 6=Sunday
df['week_of_year'] = df['datetime'].dt.isocalendar().week

# 2. Create temporal features like weekend check based on day_of_week
df['is_weekend'] = df['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)  # 1 if weekend, else 0

# 3. Creating interaction features (example)
df['duration_per_hour'] = df['duration (seconds)'] / 3600  # Convert duration from seconds to hours

# 4. Geospatial features: Calculating distance from a central point (for example, from (0,0))
central_point = (0, 0)  # Example, you can choose any location like (40.7128, -74.0060) for NYC

# Calculate distance for each observation
df['distance_from_origin'] = df.apply(lambda row: geodesic(central_point, (row['latitude'], row['longitude'])).km, axis=1)

# 5. Encoding categorical features using Label Encoding
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

# Apply label encoding to each categorical column
categorical_columns = ['state', 'country', 'shape', 'comments']
for col in categorical_columns:
    df[col] = label_encoder.fit_transform(df[col])

# 6. Feature scaling (for numerical columns)
numerical_columns = ['duration (seconds)', 'duration (hours/min)', 'latitude', 'longitude', 'distance_from_origin', 'duration_per_hour']
scaler = StandardScaler()
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

# 7. Dropping unnecessary columns (if any)
df.drop(columns=["date posted"], inplace=True)  # Dropping date posted for simplicity

# 8. Checking the final feature set
print(df.head())
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder
```

```python
cat_cols = ["city", "state", "country", "shape"]
encoder  = LabelEncoder()

for i in cat_cols:
    df[i] = encoder.fit_transform(df[i])

num_cols = ["latitude", "longitude", "duration (seconds)"]    

X = df[num_cols]


scaler = StandardScaler()

scaled_data = scaler.fit_transform(X)

pca = PCA(n_components = 2)
df_pca  = pca.fit_transform(scaled_data)

df_pca = pd.DataFrame(df_pca, columns = ["PCA1", "PCA2"])
df_pca
```
